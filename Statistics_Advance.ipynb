#Q1. What is a random variable in probability theory?
A random variable is a numerical value assigned to the outcomes of a random experiment.

Example: In a coin toss, let Heads = 1 and Tails = 0. The result becomes a random variable.

#Q2. What are the types of random variables?
Discrete Random Variable: Takes countable values (e.g., number of heads).

Continuous Random Variable: Takes uncountably infinite values over an interval (e.g., height, weight).

#Q3. What is the difference between discrete and continuous distributions?
Discrete Distribution: Deals with countable outcomes (e.g., binomial).

Continuous Distribution: Deals with outcomes over a continuous range (e.g., normal distribution).
