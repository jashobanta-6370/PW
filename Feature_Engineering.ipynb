#1. What is a parameter?
A parameter is a variable in a machine learning model that the algorithm learns from data during training (e.g., weights in linear regression).

#2. What is correlation?
Correlation measures the strength and direction of the linear relationship between two variables.

What does negative correlation mean?
It means as one variable increases, the other decreases. Example: More time studying → fewer errors on a test.

#4. How does loss value help in determining whether the model is good or not?
The loss value quantifies the difference between predicted and actual values. A low loss indicates better model performance.

#5. What are continuous and categorical variables?

Continuous: Numeric with infinite values (e.g., height, age)

Categorical: Discrete labels (e.g., gender, country)

#6. How do we handle categorical variables in Machine Learning? What are the common techniques?
Techniques:

Label Encoding: Assigns numbers to categories

One-Hot Encoding: Converts categories to binary columns

#15. What is causation? Explain difference between correlation and causation with an example.
Causation implies that one variable directly affects another.
Correlation shows relationship but not cause-effect.
Example: Ice cream sales ↑ and drowning ↑ in summer. Correlated, but ice cream doesn’t cause drowning.

#21. What is feature scaling? How does it help in Machine Learning?
Feature scaling standardizes data ranges.
It improves performance and convergence speed, especially in algorithms like KNN or SVM.

#22. How do we perform scaling in Python?
